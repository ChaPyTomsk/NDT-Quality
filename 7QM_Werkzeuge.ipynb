# Die sieben Qualitätswerkzeuge (Seven Quality Tools)
sind grundlegende Instrumente zur Qualitätskontrolle und -verbesserung. Sie wurden ursprünglich in Japan populär und von Kaoru Ishikawa als Werkzeuge beschrieben, die auch von Menschen ohne tiefgehende Statistikkenntnisse genutzt werden können, um Qualitätsprobleme zu analysieren und zu lösen.

### Hier sind die sieben Werkzeuge mit einer kurzen Beschreibung:

**1. Pareto-Diagramm**

***Zweck:*** Visualisierung der wichtigsten Probleme in einer Rangfolge, um Prioritäten zu setzen.

***Beschreibung:*** Ein Balkendiagramm, das Daten nach ihrer Bedeutung oder Häufigkeit ordnet. Es basiert auf dem Pareto-Prinzip (80/20-Regel), das besagt, dass 80 % der Ergebnisse durch 20 % der Ursachen entstehen.

***Anwendung:*** Identifikation der kritischsten Probleme in Prozessen.


**2. Ursache-Wirkungs-Diagramm (Ishikawa-Diagramm/Fischgräten-Diagramm)**

Zweck: Analyse der Ursachen für ein bestimmtes Problem.
Beschreibung: Ein Diagramm, das mögliche Ursachen eines Problems strukturiert darstellt, oft in Kategorien wie Mensch, Maschine, Material und Methode.
Anwendung: Ursachenanalyse für Probleme wie Produktionsfehler.
3. Histogramm
Zweck: Visualisierung der Verteilung von Daten.
Beschreibung: Ein Säulendiagramm, das zeigt, wie oft verschiedene Werte in einem Datensatz auftreten.
Anwendung: Analyse von Datenverteilungen, z. B. zur Erkennung von Abweichungen.
4. Kontrollkarte
Zweck: Überwachung von Prozessen über die Zeit, um Abweichungen zu erkennen.
Beschreibung: Ein Diagramm, das Datenpunkte über einen bestimmten Zeitraum darstellt, zusammen mit einer mittleren Linie und Kontrollgrenzen.
Anwendung: Prozesskontrolle und Erkennung von Trends oder Anomalien.
5. Streudiagramm
Zweck: Identifikation von Beziehungen zwischen zwei Variablen.
Beschreibung: Ein Diagramm, bei dem Punkte die Werte zweier Variablen darstellen. Ein Muster kann auf eine Korrelation hinweisen.
Anwendung: Analyse, ob und wie Variablen wie Temperatur und Produktqualität zusammenhängen.
6. Flussdiagramm
Zweck: Visualisierung eines Prozesses, um ihn besser zu verstehen und zu analysieren.
Beschreibung: Ein schematisches Diagramm, das die einzelnen Schritte eines Prozesses zeigt.
Anwendung: Identifikation von Engpässen oder ineffizienten Schritten in einem Prozess.
7. Checkliste (Prüfliste)
Zweck: Sicherstellung, dass alle notwendigen Aufgaben erledigt wurden.
Beschreibung: Eine einfache Liste von Punkten, die abgehakt werden können.
Anwendung: Fehlervermeidung und Überprüfung, z. B. bei Inspektionen.
Fazit:
Diese Werkzeuge sind vielseitig einsetzbar, um Prozesse zu analysieren, Probleme zu identifizieren und Verbesserungen umzusetzen. Sie sind besonders effektiv in der Fertigung, im Projektmanagement und in der Qualitätssicherung. Durch die Verwendung dieser Instrumente können Unternehmen fundierte Entscheidungen treffen und kontinuierlich ihre Prozesse verbessern.

**1. Pareto-Diagramm**

***Zweck:*** Visualisierung der wichtigsten Probleme in einer Rangfolge, um Prioritäten zu setzen.

***Beschreibung:*** Ein Balkendiagramm, das Daten nach ihrer Bedeutung oder Häufigkeit ordnet. Es basiert auf dem Pareto-Prinzip (80/20-Regel), das besagt, dass 80 % der Ergebnisse durch 20 % der Ursachen entstehen.

***Anwendung:*** Identifikation der kritischsten Probleme in Prozessen.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Beispiel-Daten
data = {'Fehler': ['A', 'B', 'C', 'D', 'E'], 'Häufigkeit': [40, 25, 15, 10, 5]}
df = pd.DataFrame(data)

# Sortieren und kumulative Häufigkeit berechnen
df = df.sort_values('Häufigkeit', ascending=False)
df['Kumulativ'] = df['Häufigkeit'].cumsum() / df['Häufigkeit'].sum()

# Pareto-Diagramm
fig, ax = plt.subplots()
ax.bar(df['Fehler'], df['Häufigkeit'], color='skyblue', label='Häufigkeit')
ax2 = ax.twinx()
ax2.plot(df['Fehler'], df['Kumulativ'], color='red', marker='o', label='Kumulativ')
ax2.axhline(0.8, color='gray', linestyle='--', label='80% Linie')

ax.set_ylabel('Häufigkeit')
ax2.set_ylabel('Kumulativer Anteil')
ax.legend(loc='center')
ax2.legend(loc='upper center')
plt.title('Pareto-Diagramm')
plt.show()
```

**2. Ursache-Wirkungs-Diagramm (Ishikawa-Diagramm/Fischgräten-Diagramm)**

Zweck: Analyse der Ursachen für ein bestimmtes Problem.
Beschreibung: Ein Diagramm, das mögliche Ursachen eines Problems strukturiert darstellt, oft in Kategorien wie Mensch, Maschine, Material und Methode.
Anwendung: Ursachenanalyse für Probleme wie Produktionsfehler.

```python
import matplotlib.pyplot as plt

# Zeichne ein einfaches Fishbone-Diagramm
plt.figure(figsize=(10, 6))
plt.annotate('Problem', xy=(0.7, 0.5), xytext=(0.9, 0.5), arrowprops=dict(arrowstyle='->'))

# Hauptkategorien
categories = ['Mensch', 'Maschine', 'Material', 'Methode']
y_positions = [0.7, 0.6, 0.4, 0.3]
for i, cat in enumerate(categories):
    plt.annotate(cat, xy=(0.3, y_positions[i]), xytext=(0.5, y_positions[i]), arrowprops=dict(arrowstyle='->'))

plt.axis('off')
plt.title('Fishbone-Diagramm')
plt.show()
```

3. Histogramm
Zweck: Visualisierung der Verteilung von Daten.
Beschreibung: Ein Säulendiagramm, das zeigt, wie oft verschiedene Werte in einem Datensatz auftreten.
Anwendung: Analyse von Datenverteilungen, z. B. zur Erkennung von Abweichungen.

```python
import matplotlib.pyplot as plt
import numpy as np

# Beispiel-Daten
data = np.random.normal(loc=50, scale=10, size=1000)

# Histogramm erstellen
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogramm')
plt.xlabel('Werte')
plt.ylabel('Häufigkeit')
plt.show()
```

4. Kontrollkarte
Zweck: Überwachung von Prozessen über die Zeit, um Abweichungen zu erkennen.
Beschreibung: Ein Diagramm, das Datenpunkte über einen bestimmten Zeitraum darstellt, zusammen mit einer mittleren Linie und Kontrollgrenzen.
Anwendung: Prozesskontrolle und Erkennung von Trends oder Anomalien.

```python
import numpy as np
import matplotlib.pyplot as plt

# Beispiel-Daten
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=30)

# Kontrollgrenzen
ucl = np.mean(data) + 3 * np.std(data)
lcl = np.mean(data) - 3 * np.std(data)

plt.plot(data, marker='o', label='Prozesswerte')
plt.axhline(np.mean(data), color='green', linestyle='--', label='Mittelwert')
plt.axhline(ucl, color='red', linestyle='--', label='Obergrenze (UCL)')
plt.axhline(lcl, color='red', linestyle='--', label='Untergrenze (LCL)')
plt.title('Kontrollkarte')
plt.xlabel('Stichprobe')
plt.ylabel('Wert')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Beispiel-Daten
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=30)

# Statistiken berechnen
mean = np.mean(data)
std_dev = np.std(data)

# Grenzen definieren
ucl = mean + 3 * std_dev  # Obergrenze (Upper Control Limit)
lcl = mean - 3 * std_dev  # Untergrenze (Lower Control Limit)
uwl = mean + 2 * std_dev  # Obere Warnungsgrenze (Upper Warning Limit)
lwl = mean - 2 * std_dev  # Untere Warnungsgrenze (Lower Warning Limit)

# Plot der Kontrollkarte
plt.figure(figsize=(10, 6))
plt.plot(data, marker='o', linestyle='-', label='Prozesswerte', color='blue')
plt.axhline(mean, color='green', linestyle='--', label='Mittelwert')
plt.axhline(ucl, color='red', linestyle='--', label='Obergrenze (UCL)')
plt.axhline(lcl, color='red', linestyle='--', label='Untergrenze (LCL)')
plt.axhline(uwl, color='orange', linestyle='--', label='Obere Warnungsgrenze (UWL)')
plt.axhline(lwl, color='orange', linestyle='--', label='Untere Warnungsgrenze (LWL)')

# Diagramm formatieren
plt.title('Kontrollkarte mit Warnungsgrenzen')
plt.xlabel('Stichprobe')
plt.ylabel('Prozesswert')
plt.legend(loc='upper center')
plt.grid(True)
plt.show()
```

5. Streudiagramm
Zweck: Identifikation von Beziehungen zwischen zwei Variablen.
Beschreibung: Ein Diagramm, bei dem Punkte die Werte zweier Variablen darstellen. Ein Muster kann auf eine Korrelation hinweisen.
Anwendung: Analyse, ob und wie Variablen wie Temperatur und Produktqualität zusammenhängen.

```python
import numpy as np
import matplotlib.pyplot as plt

# Beispiel-Daten
x = np.random.rand(50) * 100
y = 2 * x + np.random.normal(0, 10, 50)

plt.scatter(x, y, color='blue', label='Datenpunkte')
plt.title('Streudiagramm')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Beispiel-Daten
np.random.seed(0)
x = np.random.rand(50) * 100
y = 2 * x + np.random.normal(0, 10, 50)

# Mittelwerte berechnen
x_mean = np.mean(x)
y_mean = np.mean(y)

# Streudiagramm
plt.scatter(x, y, color='blue', label='Datenpunkte')

# Mittellinien hinzufügen
plt.axvline(x=x_mean, color='red', linestyle='--', label='Mittelwert X')
plt.axhline(y=y_mean, color='green', linestyle='--', label='Mittelwert Y')

# Lineare Regression berechnen
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Werte für die Korrelationslinie
regression_line = slope * x + intercept

# Korrelationslinie hinzufügen
plt.plot(x, regression_line, color='red', label=f'Korrelationslinie (R²={r_value**2:.2f})')

# Diagramm formatieren
plt.title('Streudiagramm mit Mittelwertslinien und Korrelationslinie')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```

6. Flussdiagramm
Zweck: Visualisierung eines Prozesses, um ihn besser zu verstehen und zu analysieren.
Beschreibung: Ein schematisches Diagramm, das die einzelnen Schritte eines Prozesses zeigt.
Anwendung: Identifikation von Engpässen oder ineffizienten Schritten in einem Prozess.

```python
from graphviz import Digraph

# Flussdiagramm erstellen
dot = Digraph(format='svg')  # Ausgabeformat auf SVG setzen (geeignet für Jupyter)

dot.attr(rankdir='LR')  # Layout von links nach rechts
dot.node('A', 'Start')
dot.node('B', 'Schritt 1: Prüfung')
dot.node('C', 'Schritt 2: Genehmigung')
dot.node('D', 'Ende')

dot.edges(['AB', 'BC', 'CD'])

# Diagramm direkt anzeigen
from IPython.display import display, SVG
display(SVG(dot.pipe(format='svg')))
```

7. Checkliste (Prüfliste)
Zweck: Sicherstellung, dass alle notwendigen Aufgaben erledigt wurden.
Beschreibung: Eine einfache Liste von Punkten, die abgehakt werden können.
Anwendung: Fehlervermeidung und Überprüfung, z. B. bei Inspektionen.

```python
import matplotlib.pyplot as plt

# Checklisten-Daten
checklist = ["Aufgabe 1: Material prüfen",
             "Aufgabe 2: Maschinen warten",
             "Aufgabe 3: Dokumentation aktualisieren",
             "Aufgabe 4: Endkontrolle durchführen"]

# Simulate a completed checklist
completed = [True, False, True, False] # Example: Items 1 and 3 are done

# Grafik-Parameter
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')  # Achsen deaktivieren

# Kopfzeile
ax.text(0.02, 1.0, "Checkliste", fontsize=16, fontweight='bold', va='top', ha='left')
ax.text(0.02, 0.95, "Datum: ______________________", fontsize=12, va='top', ha='left')
ax.text(0.02, 0.9, "Name: _______________________", fontsize=12, va='top', ha='left')

# Checklist Items
y_pos = 0.8
for i, task in enumerate(checklist):
    marker = '[✓]' if completed[i] else '[ ]' # Use check mark or empty box
    ax.text(0.02, y_pos, f"{marker}  {task}", fontsize=12, va='top', ha='left')
    y_pos -= 0.08
# Fußzeile
ax.text(0.02, y_pos - 0.1, "Unterschrift: ____________________________", fontsize=12, va='top', ha='left')
# Anzeige
plt.show()
```
